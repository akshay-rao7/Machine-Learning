```python
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn import svm

digits = datasets.load_digits()
```

```python
plt.gray() 
plt.matshow(digits.images[0]) 
plt.show() 
```

```python
print(digits.data)
```

```python
print(digits.target)
```

```python
clf = svm.SVC(gamma=0.001)
```

```python
X,y = digits.data[:-10], digits.target[:-10]
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.30, random_state=2)
```

```python
clf.fit(X_train,y_train)
```

```python
score = clf.score(X_test,y_test)
score
```

```python
%matplotlib inline
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.decomposition import PCA
```

```python
pca = PCA(3)  # project from 64 to 2 dimensions
projected = pca.fit_transform(digits.data)
print(digits.data.shape)
print(projected.shape)
```

```python
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import

import matplotlib.pyplot as plt
import numpy as np

# Fixing random state for reproducibility
np.random.seed(19680801)


def randrange(n, vmin, vmax):
    '''
    Helper function to make an array of random numbers having shape (n, )
    with each number distributed Uniform(vmin, vmax).
    '''
    return (vmax - vmin)*np.random.rand(n) + vmin

fig = plt.figure(figsize=(10,10))

ax = fig.add_subplot(111, projection='3d')


#ax.set_xlabel('X Label')
#ax.set_ylabel('Y Label')
#ax.set_zlabel('Z Label')

ax.set_xlabel('Component1')
ax.set_ylabel('Component2')
ax.set_zlabel('Component3')

ax.scatter(projected[:, 0], projected[:, 1], projected[:, 2], c=digits.target, edgecolor='none', alpha=0.5,cmap=plt.cm.get_cmap('Spectral', 10))



#plt.xlabel('component 1')
#plt.ylabel('component 2')
# plt.zlabel('component 3')
#plt.colorbar()

plt.show()
```
